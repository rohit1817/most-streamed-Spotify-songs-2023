```python
import pandas as pd

df = pd.read_csv('/content/dataset.csv', encoding = 'ISO-8859-1')
```

Check for missing values

```python
nan_values = df.isna()
any_missing_values = nan_values.any().any()
any_missing_values_in_column = nan_values.any()
missing_value_count = df.isnull().sum()
for column, has_missing in any_missing_values_in_column.items():
    if has_missing:
        count = missing_value_count[column]
        print(f" Column '{column}' has {count} missing values.")

print("\nMissing Values in the Entire DataFrame?")
print(any_missing_values)

print("\nMissing Values in Each Column?")
print(any_missing_values_in_column)

print("\nMissing Value Counts in Each Column:")
print(missing_value_count)
```

```python
#Dropping rows with missing values
df.dropna(inplace=True)
```

```python
#Changing the streams coloumn to numeric
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
```

```python
df.info()
```

```python
#Renaming some columns for better understanding

column_name_mapping = {
    'artist(s)_name': 'artist_name',
    'danceability_%': 'danceability',
    'valence_%': 'valence',
    'energy_%': 'energy',
    'acousticness_%': 'acousticness',
    'instrumentalness_%': 'instrumentalness',
    'liveness_%': 'liveness',
    'speechiness_%': 'speechiness'
}

# Use the rename method to update column names
df.rename(columns=column_name_mapping, inplace=True)

# Check the updated column names
print(df.columns)
```

```python
df.head()
```

```python
#Replacing the ',' with 'and' in the case of multiple artist names and removing non english characters from artist names column

df['artist_name'] = df['artist_name'].str.replace(',', ' and').str.replace('[^a-zA-Z\s]', '', regex=True)
```

```python
df.to_csv('new_dataset.csv', index=False)
```
